## Pandas

O `pandas` é uma biblioteca essencial para análise de dados em Python. Ele fornece estruturas de dados como `DataFrame` e `Series`, que facilitam a manipulação e análise de dados tabulares. É como ter uma planilha do Excel! Cada `DataFrame` é como uma tabela, onde você pode realizar operações como filtragem, agrupamento e agregação de dados.

Vamos importar o `pandas` e criar um `DataFrame` simples para ilustrar:

```python
import pandas as pd
```

Nessa etapa, importamos a biblioteca `pandas` e a renomeamos como `pd` para facilitar o uso. Agora, vamos criar um `DataFrame` com alguns dados de exemplo:

```python
data = {
    'Nome': ['Alice', 'Bob', 'Charlie'],
    'Idade': [25, 30, 35],
    'Cidade': ['Porto Alegre', 'João Pessoa', 'Brasília']
}
df = pd.DataFrame(data)
```

- `data`: Um **dicionário** onde as *chaves* são os **nomes das colunas** e os *valores* são **listas de dados** correspondentes a cada coluna.
- `pd.DataFrame(data)`: Cria um `DataFrame` a partir do dicionário `data`.

A tabela equivalente a esse `DataFrame` seria:

| Nome    | Idade | Cidade        |
|---------|-------|---------------|
| Alice   | 25    | Porto Alegre  |
| Bob     | 30    | João Pessoa   |
| Charlie | 35    | Brasília      |


````python
import pandas as pd 

data = {
    "Nome": ["Alice", "Bob", "Charlie"],
    "Idade": [25, 30, 35],
    "Cidade": ['Porto Alegre', 'João Pessoa', 'Brasília']
}
df = pd.DataFrame(data)
print(df)
````
